# Multicollinearity in Linear Regression

## One without Multicollinearity

```python
import pandas as pd
import statsmodels.api as sm
```

```python
df_adv = pd.read_csv('Advertising.csv',index_col=0)
```

```python
df_adv
```

```python
X = df_adv[['TV','radio','newspaper']]
y = df_adv['sales']
df_adv.head()
```

```python
# Fit the OLS model with interecept on TV and Radio
model = sm.OLS(y,X).fit()
```

```python
model.summary()
```

```python
import matplotlib.pyplot as plt
X.iloc[:,1:].corr()
```

## One with Multi-Collinearity

```python
df_salary =pd.read_csv('Salary_Data.csv')
```

```python
df_salary.head()
```

```python
X = df_salary[['YearsExperience','Age']]
y = df_salary['Salary']
```

```python
## Fit OLS Model
X=sm.add_constant(X)
model = sm.OLS(y,X).fit()
```

```python
model.summary()
```

```python
X.iloc[:,1:].corr()
```
